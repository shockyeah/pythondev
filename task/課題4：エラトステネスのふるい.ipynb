# 課題4：エラトステネスのふるい

1から100までの素数を見つけるプログラムです。

## アルゴリズム
1. 1は素数ではないため除外
2. 2を除いた2の倍数を除外
3. 3を除いた3の倍数を除外
4. 5を除いた5の倍数を除外
5. 7を除いた7の倍数を除外
6. 残った数が素数

```python
# 初期化: 1から100までの数を候補として用意
# is_prime[i]がTrueなら、iは素数候補
is_prime = [True] * 101  # インデックス0-100まで用意（0は使わない）

print("初期状態: 1から100までの数を候補として用意しました")
print(f"候補数: {sum(is_prime[1:])}個")
```

```python
# ステップ1: 1を除外（1は素数ではない）
is_prime[1] = False

print("ステップ1: 1を除外しました")
print(f"残り候補数: {sum(is_prime[1:])}個")
```

```python
# ステップ2: 2を除いた2の倍数を除外
for i in range(4, 101, 2):  # 4, 6, 8, 10, ... 100
    is_prime[i] = False

print("ステップ2: 2を除いた2の倍数を除外しました")
print(f"残り候補数: {sum(is_prime[1:])}個")
print("\n現在の候補:")
candidates = [i for i in range(1, 101) if is_prime[i]]
print(candidates)
```

```python
# ステップ3: 3を除いた3の倍数を除外
for i in range(6, 101, 3):  # 6, 9, 12, 15, ... (3の倍数で3より大きい)
    is_prime[i] = False

print("ステップ3: 3を除いた3の倍数を除外しました")
print(f"残り候補数: {sum(is_prime[1:])}個")
print("\n現在の候補:")
candidates = [i for i in range(1, 101) if is_prime[i]]
print(candidates)
```

```python
# ステップ4: 5を除いた5の倍数を除外
for i in range(10, 101, 5):  # 10, 15, 20, 25, ... (5の倍数で5より大きい)
    is_prime[i] = False

print("ステップ4: 5を除いた5の倍数を除外しました")
print(f"残り候補数: {sum(is_prime[1:])}個")
print("\n現在の候補:")
candidates = [i for i in range(1, 101) if is_prime[i]]
print(candidates)
```

```python
# ステップ5: 7を除いた7の倍数を除外
for i in range(14, 101, 7):  # 14, 21, 28, 35, ... (7の倍数で7より大きい)
    is_prime[i] = False

print("ステップ5: 7を除いた7の倍数を除外しました")
print(f"残り候補数: {sum(is_prime[1:])}個")
print("\n現在の候補:")
candidates = [i for i in range(1, 101) if is_prime[i]]
print(candidates)
```

```python
# 最終結果: 1から100までの素数を表示
print("=" * 50)
print("最終結果: 1から100までの素数")
print("=" * 50)

for i in range(1, 101):
    if is_prime[i]:
        print(i)

print("\n" + "=" * 50)
print(f"素数の個数: {sum(is_prime[1:])}個")
```
