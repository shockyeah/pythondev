# 課題3：平均点を求めるプログラム

## 仕様
* 無限ループを作成し、その中でテストの点数（整数値）を入力します
* 入力した点数の件数（人数）をカウントします
* 点数の合計値を繰り返しごとに加算します
* -1 と入力されたら、繰り返し処理を終了します
* 平均点は「点数の合計値 / 人数」で求めます
* 最後に「○○人のテストの平均点は△△点です」と表示します
* 平均点は小数第2位を四捨五入し、小数第1位までを表示します

## プログラムの流れ
1. **変数の初期化**: 合計点と人数カウンタを0で初期化
2. **ループ処理**: 点数を入力し、-1が入力されるまで繰り返す
3. **平均計算**: 合計点 ÷ 人数で平均を計算
4. **結果表示**: 小数第1位まで表示

```python
# 【処理1】変数の初期化
total_score = 0  # 点数の合計値
count = 0        # 人数のカウンタ
```

```python
# 【処理2】ループ処理（入力・カウント・合計）
while True:
    # 点数を入力
    score = int(input("点数を入力してください："))
    
    # -1が入力されたらループを終了
    if score == -1:
        break
    
    # 点数を合計値に加算
    total_score += score
    
    # 人数をカウント
    count += 1
```

```python
# 【処理3】平均計算と結果表示
# 平均点を計算（小数第2位を四捨五入）
average = round(total_score / count, 1)

# 結果を表示（小数第1位まで表示）
print(f"{count}人のテストの平均点は{average:.1f}点です")
```
