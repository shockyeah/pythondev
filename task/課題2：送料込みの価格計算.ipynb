# 課題2：送料込みの価格計算

税抜価格から税込価格を計算し、送料を含めた最終価格を求めるプログラムです。

## 仕様
- 税率：10%（税抜価格 × 1.1）
- 小数点以下は四捨五入
- 税込価格が2000円以上：送料無料
- 税込価格が2000円未満：送料350円

## ステップ1：税抜価格の入力

```python
# 税抜価格の入力を受け付ける
price_before_tax = int(input("税抜価格を入力してください："))
```

## ステップ2：税込価格の計算

```python
# 税込価格を計算（10%の税率）
# 小数点以下を四捨五入するためround()関数を使用
price_with_tax = round(price_before_tax * 1.1)
```

## ステップ3：送料の判定と加算

```python
# 税込価格によって送料を判定
if price_with_tax >= 2000:
    # 2000円以上の場合：送料無料
    print("送料は無料です")
    final_price = price_with_tax
else:
    # 2000円未満の場合：送料350円
    print("送料として350円かかります")
    final_price = price_with_tax + 350
```

## ステップ4：最終価格の表示

```python
# 送料込みの最終価格を表示
print(f"送料込みの価格は{final_price}円です。")
```
