# 課題8：単語内の文字数カウント

英単語を入力して、各文字の出現回数をカウントするプログラムです。

## 1. 英単語を格納するための空のリストを作成

```python
# 英単語を格納するための空のリストを作成
word_list = []
print("英単語リストを初期化しました:", word_list)
```

## 2. アルファベット小文字の各文字をキーとする辞書を作成（値は0で初期化）

```python
# アルファベット小文字のa-zをキーとし、値を0で初期化した辞書を作成
char_count = {
    'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0, 'h':0, 'i':0, 'j':0,
    'k':0, 'l':0, 'm':0, 'n':0, 'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0,
    'u':0, 'v':0, 'w':0, 'x':0, 'y':0, 'z':0
}
print("文字カウント用辞書を初期化しました:", char_count)
```

## 3. 英単語の入力を受け付け、文字数をカウント

```python
# 英単語の入力ループ
while True:
    # キーボードから英単語を入力
    word = input("英単語を入力してください：")
    
    # 未入力（空文字列）の場合はループを終了
    if word == "":
        break
    
    # 入力された英単語をリストに追加
    word_list.append(word)
    
    # 単語を小文字に変換して、1文字ずつ確認
    for char in word.lower():
        # 英字（a-z）のみをカウント対象とする
        if char in char_count:
            char_count[char] += 1
```

## 4. 結果の表示

```python
# 入力された英単語をABC順（昇順）に並べ替えて表示
sorted_word_list = sorted(word_list, key=str.lower)
print("入力した英単語：" + str(sorted_word_list))

# 1回以上出現した文字について「○が△個ありました」を表示
for char, count in char_count.items():
    if count > 0:
        print(f"{char}が{count}個ありました")
```
